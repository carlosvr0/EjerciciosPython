#### MENOR DE DOS PARES: Escriba una función que devuelva el menor de dos números dados *si* ambos números son pares, pero devuelva el mayor si uno o ambos números son impares
    lesser_of_two_evens(2,4) --> 2
    lesser_of_two_evens(2,5) --> 5

```python
def lesser_of_two_evens(a,b):
     if a%2==0 and b%2==0: # Si ambos son pares, devuelve el menor
        return min(a,b)
     else:
        return max(a,b) # i unoi o ambos son impares, devuelve el mayor
```

```python
# Revisión
lesser_of_two_evens(2,4)
```

```python
# Revisión
lesser_of_two_evens(2,5)
```

#### GALLETAS DE ANIMALES: Escribir una función que toma una cadena de dos palabras y devuelve True si ambas palabras comienzan con la misma letra
    animal_crackers('Ciudad Calor') --> True
    animal_crackers('Lugar Frio') --> False

```python
def animal_crackers(text):
  p1, p2 = text.split() # divide el texto en dos palabras
  return p1[0].lower() == p2[0].lower() #compara la 1ra letra de ambas palabras en minuscula
```

```python
# Revisión
animal_crackers('Ciudad Calor')
```

```python
# Revisión
animal_crackers('Lugar Frio')
```

#### HACE VEINTE: dados dos enteros, devuelve True si la suma de los enteros es 20 o si uno de los enteros es 20. De lo contrario, devuelve False

    makes_twenty(20,10) --> True
    makes_twenty(12,8) --> True
    makes_twenty(2,3) --> False

```python
def makes_twenty(n1,n2):
  return (n1 + n2 == 20) or (n1==20) or (n2==20)
```

```python
makes_twenty(12,8)
```

#### MAYÚSCULAS: Escribir una función que haga mayúscula la primera y cuarta letra de una cadena de texto
     
    old_macdonald('macdonald') --> MacDonald
    
Note: `'macdonald'.capitalize()` returns `'Macdonald'`

```python
def old_macdonald(name):
  return name[:3].capitalize() + name[3:].capitalize() # Convierte en mayúscula la primera y la cuarta letra

old_macdonald('carVille')
```

#### REVERSA: Dada una oración, retornar la misma oración pero en reversa

    master_yoda('I am home') --> 'home am I'
    master_yoda('We are ready') --> 'ready are We'
    
Note: Revisar método .join()

    >>> "--".join(['a','b','c'])
    >>> 'a--b--c'

Resultado de .join():

    >>> " ".join(['Hello','world'])
    >>> "Hello world"

```python
def master_yoda(text):
    return " ".join(reversed(text.split()))   # se invierte la lista de palabras sin usar slicing
```

#### PROBLEMA 33:

Dada una lista de enteros, retornar True si el arreglo tinene un  3  seguido por un  3 .

    has_33([1, 3, 3]) → True
    has_33([1, 3, 1, 3]) → False
    has_33([3, 1, 3]) → False

```python
def has_33(nums):
    for i in range(len(nums)-1): # Recorre la lista
      if nums[i] == 3 and nums[i+1] == 3: # Si hay un 3 seguido de otro 3
        return True
    return False # no hay dos 3 seguidos
```

```python
# Revisión
has_33([1, 3, 3])
```

```python
# Revisión
has_33([1, 3, 1, 3])
```

```python
# Revisión
has_33([3, 1, 3])
```

#### REPLICADOR: Dada una cadena, devuelve una cadena donde por cada carácter en el original hay tres caracteres REPETIDOS
    paper_doll('Hello') --> 'HHHeeellllllooo'
    paper_doll('Mississippi') --> 'MMMiiissssssiiippppppiii'

```python
def paper_doll(text):
    return "".join(letra * 3 for letra in text) # Repite cada carácter 3 veces y los une en una sola cadena
```

```python
# Check
paper_doll('Hello')
```

```python
# Check
paper_doll('Mississippi')
```

#### BLACKJACK: Dados tres enteros entre 1 y 11, si su suma es menor o igual a 21, devuelve su suma. Si su suma excede 21 *y* hay un once, reduce la suma total en 10. Finalmente, si la suma (incluso después del ajuste) excede 21, devuelve 'BUST'
    blackjack(5,6,7) --> 18
    blackjack(9,9,9) --> 'BUST'
    blackjack(9,9,11) --> 19

```python
def blackjack(a,b,c):
    sumNum = a+b+c #suma inicial
    if sumNum <= 21:  #si es menor igual a 21, devuelve la suma
      return sumNum
    elif sumNum >21 and 11 in (a,b,c): #supera 21 y hay un 11 en los números, se resta 10
      sumNum -= 10
      return sumNum if sumNum <= 21 else 'BUST'
    else:
      return 'BUST' #si no se cumple nada de lo anterior
```

```python
# Revisión
blackjack(5,6,7)
```

```python
# Revisión
blackjack(9,9,9)
```

```python
# Revisión
blackjack(9,9,11)
```
